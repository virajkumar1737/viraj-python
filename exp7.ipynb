1. Write a Python function to find the maximum and minimum numbers from a
sequence of numbers. (Note: Do not use built-in functions.)

```python
def find_max_min(nums):
    maximum=nums[0]
    minimum=nums[0]

    for n in nums:
        if n>maximum:
            maximum=n
        if n<minimum:
            minimum=n
    
    return maximum,minimum

numbers = list(map(int,input("netr number:").split()))

maxi,mini=find_max_min(numbers)
print("maximum =",maxi)
print("minimum =",mini)
```

Explanation: the function compares each number with currento max and min values. largest and smallest values are updated manully without using built in function.

2. Write a Python function that takes a positive integer and returns the sum of the
cube of all the positive integers smaller than the specified number.

```python
def sum_of_cubes(n):
    total = 0
    for i in range(1,n):
        total = total+ (i**3)
    return total

num = int(input("enter a number:"))
print("sum o cubes =",sum_of_cubes(num))
```

Explanation: the function calculates cubes of all number smaller than the given number. it adds them together and returns the total sum.

3. Write a Python function to print 1 to n using recursion. (Note: Do not use loop)

```python
def print_numbers(n):
    if n==0:
        return
    print_numbers(n-1)
    print(n)

num = int(input("enter a number:"))
print_numbers(num)
```

Explanation: the function calls itself to reach 0 first. then numbers are printed while returning back from recursion.

4. Write a recursive function to print Fibonacci series upto n terms.

```python
def fib(n):
    if n<=1:
        return n
    return fib(n-1) + fib(n-2)

def series(n):
    if n==0:
        return
    series(n-1)
    print(fib(n-1),end="")

num = int(input("enter number of terms:"))
series(num)
```

Explanation: fib() finds fibonacci numbers using recursion. series() prints first n terms recursively without using loops.

5. Write a lambda function to find volume of cone.

```python
volume = lambda r,h:(1/3)* 3.14*r*r*h

r = float(input("enter radius:"))
h = float(input("enter height:"))

print("volume of cone =",volume(r,h))
```

Explanation: a lambda function is used to calulate volume using formula.

6. Write a lambda function which gives tuple of max and min from a list.
Sample input: [10, 6, 8, 90, 12, 56]
Sample output: (90,6)

```python
max_min=lambda lst:(max(lst),min(lst))

nums=list(map(int,input("Enter numbers: ").split()))

print(max_min(nums))
```

Explanation: the lambda function returns a tuple containing maximum and minimum values. it uses max() and min() on the given list
