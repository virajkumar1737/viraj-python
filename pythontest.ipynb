```python
def top_k(items, k):
    sorted_items = sorted(items,key=lambda x: x[1],reverse=True)
    return items[:k]
```

```python
def count_by_category(pairs):
    d = {}
    for cat, word in pairs:
        d.setdefault(cat, {})
        d[cat].setdefault(word, 0)
        d[cat][word] += 1
    return d
```

```python
def missing_number(nums):
    n = len(nums)
    total = n * (n + 1) // 2
    return int(total - sum(nums))
```

```python
def add_unique(x, seen=None):
    if seen is None:
        seen = set()
        if x is not seen:
            seen.add(x)
    return list(seen)
```

```python
def moving_average(stream, w):
    result = []
    window_sum = 0

    for i in range(len(stream)):
        window_sum += stream[i]

        if i >= w:
            window_sum -= stream[i - w]

            current_window_size = min(i + 1,w)

            result.append(window_sum / current_window_size)
            return result
    

```
