#### 1. Check whether a given number is divisible by 3 and 5 both.

```python
num = int(input("enter a number:"))
if num % 3 == 0 and num % 5 == 0:
    print("number is divisible by both 3 and 5")
else:
        print("number is not divisible by both 3 and 5")
```

explaintion: the program checks the remainder when the number is divided by 3 and 5 if both the remainders is 0 then the number is divisible by both

#### 2 . Check whether a given number is multiple of five or not.

#### 3.Find the greatest among two numbers. If numbers are equal than print “numbers are equal”.

```python
a = int(input("enter first number:"))
b = int(input("enter second number:"))

if a > b:
    print("greatest number is:",a)
elif b > a:
    print("greatest number is:",b)
else:
    print("numbers are eqaul")
```
